```python
from urllib.parse import parse_qs

my_values = parse_qs('http://example.com/?a=1&b=2&c=3', keep_blank_values=True)
#print(repr(my_values))

red = my_values.get('a',[''])[0]
print(f'Red: {red}')
for val in my_values:
    print(f'Value: {val}')
```

```python
from random import randint
random_bits = 0
for i in range(10):
    print(f'Random number: {randint(0, 1)}')
    
for i in range(32):
    random_bits = (random_bits << 1) | randint(0, 1)
print(f'Random bits: {random_bits:032b}')
```

```python
import itertools

names = ['peter', 'sally', 'nick']
counts = [0, 1, 2]
names.append('odysseas')
print(names)
for name, count in itertools.zip_longest(names, counts):
    print(f'{name} has {count} items')
```

```python
car_values = [2,4,7,11,3,20,5]
car_sorted = sorted(car_values, reverse=True)
oldest, second_oldest, *rest = car_sorted
print(f'Oldest: {oldest}')
print(f'Second oldest: {second_oldest}')
print(f'Rest: {rest}')
```

```python
class Tool():
    def __init__(self, name, weight=0):
        self.weight = weight
        self.name = name

    def __repr__(self):
        return f'Tool({self.name}) has weight {self.weight})\n'

tools = [Tool('hammer', 2), Tool('screwdriver', 5), Tool('wrench', 2), Tool('saw', 40)]
tools.sort(key=lambda x: (x.name, x.weight), reverse=True)
print(f'Tools: {tools}')
```
